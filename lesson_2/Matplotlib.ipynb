# Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color='#00c403')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title('График f(t)', fontsize=15, fontweight='bold')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

# Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y_1 = x**2
y_1
```

```python
y_2 = 2 * x + 0.5
y_2
```

```python
y_3 = -3 * x - 1.5
y_3
```

```python
y_4 = np.sin(x)
y_4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()
ax_1.plot(x, y_1)
ax_1.set_title('График y_1')
ax_1.set_xlim([-5, 5])

ax_2.plot(x, y_2)
ax_2.set_title('График y_2')

ax_3.plot(x, y_3)
ax_3.set_title('График y_3')

ax_4.plot(x, y_4)
ax_4.set_title('График y_4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
